###   Functions for dashboard

```python
# Function to create Passenger Journeys by Mode of Transport Dublin plot
def create_passenger_journeys_plot():
    number_passenger_group = passenger_journeys.groupby('Mode of Transport')['VALUE'].sum().reset_index()
    fig = px.bar(number_passenger_group, x='Mode of Transport', y='VALUE', color='Mode of Transport')
    fig.update_layout(title='Passenger Journeys by Mode of Transport Dublin')
    return fig

# Function to create Number of passengers over Time in Dublin plot
def create_passenger_over_time_plot():
    fig = px.line(passenger_journeys, x='Year', y='VALUE', line_group='Year')
    fig.update_layout(title='Number of passengers over Time in Dublin')
    return fig

# Function to create Number of passengers over Time in Dublin (Weekly) plot
def create_passenger_over_time_weekly_plot():
    fig = px.line(passenger_journeys, x='Week', y='VALUE', color='Year')
    fig.update_layout(title='Number of passengers over Time in Dublin (Weekly)')
    return fig
```

```python
# Function to create Number of Passengers over Time in London plot
def create_number_of_passengers_over_time_london_plot():
    fig, ax = plt.subplots(figsize=(15, 5))
    sns.lineplot(data=Type_of_Transport_london_melt, x="Year", y="Value_in_millions", color="b", ax=ax)
    ax.set_title("Number of Passengers over Time in London")
    ax.set_xlabel("Year")
    ax.set_ylabel("Number of Passengers (Millions)")
    ax.grid(True)
    return fig

# Function to create Passenger Journeys by Mode of Transport London plot
def create_passenger_journeys_by_mode_london_plot():
    number_passenger_group_london = Type_of_Transport_london_melt.groupby('Mode of Transport')['Value_in_millions'].sum().reset_index()
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(number_passenger_group_london['Mode of Transport'], number_passenger_group_london['Value_in_millions'], color='r')
    ax.set_xlabel('Mode of Transport')
    ax.set_ylabel('Value_in_millions')
    ax.set_title('Passenger Journeys by Mode of Transport')
    ax.set_xticklabels(number_passenger_group_london['Mode of Transport'], rotation=45, ha='right')
    plt.tight_layout()
    return fig

# Function to create Number of Passengers over Time in London plot (Weekly)
def create_number_of_passengers_over_time_weekly_plot():
    fig, ax = plt.subplots(figsize=(15, 5))
    sns.lineplot(data=Type_of_Transport_london_melt, x="Reporting Period", y="Value_in_millions", hue="Year", ax=ax)
    ax.set_title("Number of Passengers over Time in London")
    ax.set_xlabel("Week")
    ax.set_ylabel("Number of Passengers")
    ax.xticks(rotation=45)
    return fig
```
